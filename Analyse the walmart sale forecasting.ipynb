# Data Analysing & Visualisation

```python
data1.plot()
data2.plot()
data3.plot()
```

```python
data1.hist(figsize=(5,10))
data2.hist(figsize=(5,10))
data3.hist(figsize=(5,10))
```

```python
dataset=data2.merge(data3,how='inner',on='Store')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.tail()
```

```python
dataset.tail()
```

```python
from datetime import datetime
```

```python
dataset['Date_x'] = pd.to_datetime(dataset['Date_x'])
print(dataset['Date_x'])
```

```python
data2['Date'] = pd.to_datetime(data2['Date'])
print(data2['Date'])
```

```python
data3['Date'] = pd.to_datetime(data3['Date'])
print(data3['Date'])
```

```python
dataset['week'] = dataset.Date_x.dt.isocalendar().week
dataset['year'] = dataset.Date_x.dt.isocalendar().year
```

```python
print(dataset['week'])
```

```python
print(dataset['year'])
```

```python
def scatter(mart,column):
     plt.figure()
     plt.scatter(data2[column], mart['Weekly_Sales'])
     plt.xlabel(column)
     plt.ylabel('Weekly_Sales')
```

```python
scatter(data2, 'Store')
```

```python
scatter(data2, 'Dept')
```

```python
scatter(data2, 'Temperature')
```

```python
weekly_sales = dataset[dataset['year'] == 2010]['Weekly_Sales_x'].groupby(dataset['week']).mean()
sb.lineplot(weekly_sales.index, weekly_sales.values,color='blue')
```

```python
df = pd.concat([data2['Type'], data2['Size']], axis=1)
```

```python
fig = sb.boxplot(x='Type', y='Size', data=df)
```

```python
plt.figure(figsize= (15, 15))
sb.barplot(weekly_sales.index, weekly_sales.values,palette='dark')
plt.title(' Weekly Sales Per Store', fontsize = 20)
plt.xlabel('Stores', fontsize = 16)
plt.ylabel('Sales ', fontsize = 16)
plt.show()
```

```python
weekly_sales_Dept = data2['Weekly_Sales'].groupby(data2['Dept']).mean()
print(weekly_sales_Dept)
```

```python
weekly_sales_Temp=data2['Weekly_Sales'].groupby(data2['Temperature']).mean()
print(weekly_sales_Temp)
```

```python
sb.set(style = 'white')
fig, ax = plt.subplots(figsize= (15, 15))
sb.heatmap(data2.corr(), cmap = 'icefire', vmax = 0.3, center = 0, square = True, linewidth= 0.5, cbar_kws = {'shrink': 0.5}, annot = True)
```
